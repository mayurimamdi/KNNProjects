```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
cols = df[['Brand', 'Category','Color','Size']]
```

```python
cols
```

```python
cols_df=pd.get_dummies(cols).astype(int)
```

```python
df=pd.concat([df,cols_df], axis=1)
```

```python
df.head()
```

```python
df.drop(columns=['User ID','Product ID', 'Product Name','Category','Brand','Color','Size'], axis=1, inplace=True)
```

```python
df
```

```python
df.shape
```

```python
X=df.drop(columns='Rating')
y=df['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rfr=RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
y_pred_rfr = rfr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred_rfr)
r2 = r2_score(y_test, y_pred_rfr)
```

```python
mse
```

```python
r2
```

```python
knn = KNeighborsRegressor()
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)
```

```python
mse_knn
```

```python
r2_knn
```

Lets try linear regression with standard scaling

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.fit_transform(X_test)

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_test_scaled)

mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
```
