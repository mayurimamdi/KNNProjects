```python
df['Rating'].unique()
```

```python
sns.countplot(x='Rating', data=df)
xticks = plt.xticks(rotation=75)
plt.show()
```

So the highes ratings got to the booksis 4.8 out of 5

```python
sns.barplot(x='Rating', y='price', data=df, ci=None)
xticks = plt.xticks(rotation=75)
plt.show()
```

```python
book_ratings=df.groupby('Book name')['Rating'].value_counts().sort_values(ascending=False)
book_ratings.head(20)
```

```python
sns.barplot(x='Book name', y='Rating', data=book_ratings.head(10).reset_index().sort_values(by='Rating', ascending=False), palette='viridis')
plt.xticks(rotation=90)
plt.show()
```
