```python
df = df.rename(columns={
    'Book name': 'title',
    'Author': 'author',
    'Rating': 'rating',
    'reviews count': 'reviews_count',
    'Print Length': 'print_length',
    'Publishing date': 'publishing_date',
    'Genre': 'genre'
})
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df = df[['title', 'author', 'genre', 'rating', 'reviews_count', 'print_length']]
```

```python
df['combined_text'] = df['title'] + " " + df['author'] + " " + df['genre']

print("Final Shape:", df.shape)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):

    text = str(text).lower()
    

    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    words = text.split()
    

    clean_words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]
    
    return " ".join(clean_words)

print("Processing book text... This might take a moment.")

df['text_clean'] = df['combined_text'].apply(preprocess_text)

print("Done!")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=3000,
    ngram_range=(1,2)
)

tfidf_matrix = tfidf.fit_transform(df['text_clean'])

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def recommend_books(title, cosine_sim=cosine_sim):
    
    if title not in indices:
        return "Book not found in dataset."
    
    idx = indices[title]
    
    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort books by similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Remove itself (first one is always same book)
    sim_scores = sim_scores[1:6]
    
    # Get book indices
    book_indices = [i[0] for i in sim_scores]
    
    return df['title'].iloc[book_indices]
```

```python
def evaluate_genre_overlap(title):
    idx = indices[title]
    original_genre = df.loc[idx, 'genre']
    
    recommendations = recommend_books(title)
    
    overlap_count = 0
    
    for book in recommendations:
        rec_genre = df[df['title'] == book]['genre'].values[0]
        if rec_genre == original_genre:
            overlap_count += 1
    
    print("Genre Match Count:", overlap_count, "/ 5")
```

```python
def evaluate_genre_overlap(title):
    
    if title not in indices:
        print("Book not found in dataset.")
        return
    
    idx = indices[title]
    original_genre = df.loc[idx, 'genre']
    
    recommendations = recommend_books(title)
    
    overlap_count = 0
    
    for book in recommendations:
        rec_genre = df[df['title'] == book]['genre'].values[0]
        
        if rec_genre == original_genre:
            overlap_count += 1
    
    print(f"Original Genre: {original_genre}")
    print(f"Genre Match Count: {overlap_count} / {len(recommendations)}")
```
